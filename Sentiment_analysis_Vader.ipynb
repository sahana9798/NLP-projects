```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('moviereviews.tsv',sep='\t')
df.head()
```

```python
df['label'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
blanks=[]
for i,lb,rv in df.itertuples():
  if rv.isspace():
    blanks.append(i)
```

```python
blanks
```

```python
df.drop(blanks,inplace=True)
```

```python
import nltk
nltk.downloader.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid= SentimentIntensityAnalyzer()
```

```python
df['scores']=df['review'].apply(lambda x: sid.polarity_scores(x))
```

```python
df.head()
```

```python
df['compound_score']=df['scores'].apply(lambda d:d['compound'])
```

```python
df.head()
```

```python
df['predicted_label']=df['compound_score'].apply(lambda x: 'pos' if x>0 else 'neg')
```

```python
df.head()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
accuracy_score(df['label'],df['predicted_label'])
```

```python
confusion_matrix(df['label'],df['predicted_label'])
```

```python
print(classification_report(df['label'],df['predicted_label']))
```
